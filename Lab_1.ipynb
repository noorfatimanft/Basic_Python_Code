1. Write a program to check if a number is even or odd.

```python
num = int(input("Enter a number: "))
print("Even" if num % 2 == 0 else "Odd")
```

num % 2 checks the remainder to decide if the number is even or odd.

2. Write a function that takes a string and returns it in uppercase.

```python
def to_upper(s):
    return s.upper()

print(to_upper("upper"))
```

.upper() converts all characters in the string to uppercase.

3. Print the first 10 natural numbers using a loop.

```python
for i in range(1, 11):
    print(i)
```

range(1, 11) generates numbers from 1 to 10.

4. Write a program to find the largest of three numbers.

```python
a, b, c = 10, 20, 5
print(max(a, b, c))
```

max() returns the biggest value among the given numbers.

5. Write a program to swap two numbers without using a third variable

```python
x = 5
y = 10
x, y = y, x
print("x = ",x, "y = ",y)
```

Python swaps values directly using tuple unpacking.

6. Take two numbers from the user and print their sum, difference, product, and quotient.

```python
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))

print("Sum:", a + b)
print("Difference:", a - b)
print("Product:", a * b)
print("Quotient:", a / b)
```

This program takes two numbers from the user, performs all basic arithmetic operations (add, subtract, multiply, divide), and prints the results.

7. Write a program that checks if a year is a leap year or not.

```python
year = int(input("Enter year: "))
print("Leap Year" if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0) else "Not Leap Year")
```

Uses the leap year rule: divisible by 4 but not 100 unless divisible by 400.

9. Write a program to remove duplicates from a list.

```python
lst = [1, 2, 2, 3, 4, 4, 5]
lst = list(set(lst))
print(lst)
```

set() removes duplicates because sets allow only unique elements.
